```python
e = 2.71828
y_scale = 1/20
x_inf_lim, x_sup_lim = -2, 2
point_count = 0
plus_count = 0 
vetor_y = []

def f(x): 
    return e**(-x **2)
    
for i in range(x_inf_lim * 10, (x_sup_lim * 10) + 1, 1): 
    point_count += int(f(i/10)/y_scale)
    plus_count += 1
    print('{}{}'.format('.' * int(f(i/10)/y_scale), '+'))  
    vetor_y.append(int(f(i/10)/y_scale))

print(plus_count, point_count)
```

```python
point_count = 0
plus_count = 0 

for i in range(max(vetor_y), -1, -1):
            
    for j in range(0, len(vetor_y)):

        if vetor_y[j] < i:
            print(' ', end='')
        
        elif vetor_y[j] > i: 
            print('.', end='')
            point_count += 1
            
        else:
            print('+', end='')
            plus_count += 1
         
        print(' ', end='') #visual            

    print('')

print(plus_count, point_count)
```

```python

# Gráfico não é fidedigno ao original, mas a ideia do plot é boa.

posicao = -1
point_count = 0
plus_count = 0 
 
for i in range(max(vetor_y), -1, -1):
    
    vetor_posicao = [a for a in vetor_y if a >= i]

    if i in vetor_posicao:
        posicao = vetor_y.index(i)

    print(' '*posicao, end = '')
    print('+', end = '')
    print('.'*len(vetor_posicao), end = '')
    print('+', end = '')
    print('') 

    plus_count  += 2  
    point_count += len(vetor_posicao)

print(plus_count, point_count)
    
```
